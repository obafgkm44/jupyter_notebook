必要なタグを抽出する

```python
import requests
from bs4 import BeautifulSoup

response = requests.get('https://toiguru.jp/toeic-vocabulary-list#smoothplay1')

soup = BeautifulSoup(response.text,'lxml')

links = soup.findAll('td')
for link in links:
    
    print(link)
```
